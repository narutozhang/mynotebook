```python
from numpy import random
import numpy as np
```

```python
X = random.rand(100)
Y = 5 * X**2 + 1.2 * X + 0.8
Z = random.randn(100)/100
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

ax=plt.subplot(111,projection='3d')
ax.scatter(X,Y,Z)
```

```python
plt.subplot(1,3,1)
plt.scatter(X, Y)
plt.subplot(1,3,2)
plt.scatter(X, Z)
plt.subplot(1,3,3)
plt.scatter(Y, Z)
```

```python
from sklearn.decomposition import PCA

pca = PCA(2)

X2 = pca.fit_transform(np.c_[X,Y,Z])
```

```python
plt.scatter(X2[:,0], X2[:,1])
```
